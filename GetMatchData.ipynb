```python
import import_ipynb
import BrowserConfig as bc
import CommonMethods as cm
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
#Function for gathering all match data
def get_match_data(driver, match_data, option_id):
    #Open window for Preliminiary Round or Final 8
    cm.wait_on_element(driver, (By.ID, "cmbCalendar"))
    cm.click_on_element(driver, driver.find_element(By.ID, option_id))
    
    #Getting all calendar dates 
    all_calendar_days = cm.wait_on_all_elements(driver, (By.XPATH, '//div[@id="tblBottoniDay"]/div'))
    
    #Going through each date
    for matchdays in range(1,len(all_calendar_days)+1):
        
        #Assuring that opens right window each iteration 
        cm.wait_on_element(driver, (By.ID, "cmbCalendar"))
        cm.click_on_element(driver, driver.find_element(By.ID, option_id))

        #Opens window of matchday
        element = cm.wait_on_element(driver, (By.XPATH, '//div[@id="tblBottoniDay"]['+str(matchdays)+']/div'))
        driver.execute_script("arguments[0].click();", element)
    
        #Getting all matches from date
        all_matches_matchday = cm.wait_on_all_elements(driver, (By.XPATH, '//div[@class="d-block d-lg-none"]/div[@class="row calendar-event"]/div[@class="col-12 col-lg-2 date"]/span'))
        
        #Going through each match
        for match in all_matches_matchday:
            try:
                #Opens window for match
                driver.execute_script("arguments[0].click();", match)
            except:
                
                #Preventing Stale Element Excpetion
                #Assures the right matchday
                cm.wait_on_element(driver, (By.ID, "cmbCalendar"))
                cm.click_on_element(driver, driver.find_element(By.ID, option_id))
                element = cm.wait_on_element(driver, (By.XPATH, '//div[@id="tblBottoniDay"]['+str(matchdays)+']/div'))
                driver.execute_script("arguments[0].click();", element)

            #Data mining for Home team
            #Adding it to DataFrame
            team1 = cm.get_text_from_element(driver,(By.XPATH, '//div[@id="tdStat1Final"]/table/tbody/tr/td[@class="headerContenutiPN_PBP"]/strong'))
            try:
                team1_data_elements = cm.wait_on_all_elements(driver, (By.XPATH, '//div[@id="tdStat1Final"]/table/tbody/tr[@class="trContenutiTR"]/td[@class="aCenter "]'))
            except:
                #Preventing Timeout Exception
                #Some insantces don't use the original XPATH
                team1_data_elements = cm.wait_on_all_elements(driver, (By.XPATH, '//div[@id="tdStat1Final"]/table/tbody/tr[@class="trContenutiToggleTR"]/td[@class="aCenter "]'))
            
            team1_data = cm.get_list_text_from_elements(driver, team1_data_elements)
            cm.add_row_data_frame(match_data, team1, team1_data)

            #Data mining for Away team
            #Adding it to DataFrame
            team2 = cm.get_text_from_element(driver,(By.XPATH, '//div[@id="tdStat2Final"]/table/tbody/tr/td[@class="headerContenutiPN_PBP"]/strong'))
            try:
                team2_data_elements = cm.wait_on_all_elements(driver, (By.XPATH, '//div[@id="tdStat2Final"]/table/tbody/tr[@class="trContenutiTR"]/td[@class="aCenter "]'))
            except:
                #Preventing Timeout Exception
                #Some insantces don't use the original XPATH
                team2_data_elements = cm.wait_on_all_elements(driver, (By.XPATH, '//div[@id="tdStat2Final"]/table/tbody/tr[@class="trContenutiToggleTR"]/td[@class="aCenter "]'))
            
            team2_data = cm.get_list_text_from_elements(driver, team2_data_elements)
            cm.add_row_data_frame(match_data, team2, team2_data)
            
            #print(match_data)
            #Goes back to previous window
            cm.click_on_element(driver, driver.find_element(By.XPATH, '//div[@id="tblTabsSotto"]/div/div[@class="col-6 col-sm-6 text-left"]/a'))
            continue
        continue
```
